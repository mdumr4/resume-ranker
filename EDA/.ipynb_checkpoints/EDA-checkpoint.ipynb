```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Data Loading and Flattening
Since the dataset has ~100k records with deeply nested structures, we will parse the JSONL file and flatten it into a Pandas DataFrame for easier analysis.

```python
def flatten_candidate(c):
    flat = {'candidate_id': c.get('candidate_id')}

    # Profile
    profile = c.get('profile', {})
    for k, v in profile.items():
        flat[f'profile_{k}'] = v

    # Redrob Signals
    signals = c.get('redrob_signals', {})
    for k, v in signals.items():
        if k == 'expected_salary_range_inr_lpa':
            flat['expected_salary_min'] = v.get('min') if isinstance(v, dict) else None
            flat['expected_salary_max'] = v.get('max') if isinstance(v, dict) else None
        elif k == 'skill_assessment_scores':
            # Store average assessment score or number of assessments taken
            scores = list(v.values()) if isinstance(v, dict) else []
            flat['signal_assessments_taken'] = len(scores)
            flat['signal_avg_assessment_score'] = sum(scores) / len(scores) if scores else None
        else:
            flat[f'signal_{k}'] = v

    # Arrays (Career, Education, Skills, Certifications, Languages)
    flat['num_jobs'] = len(c.get('career_history', []))
    flat['num_degrees'] = len(c.get('education', []))
    flat['num_skills'] = len(c.get('skills', []))
    flat['num_certifications'] = len(c.get('certifications', []))
    flat['num_languages'] = len(c.get('languages', []))

    # Skill specific aggregations
    skills = c.get('skills', [])
    flat['avg_skill_endorsements'] = sum(s.get('endorsements', 0) for s in skills) / len(skills) if skills else 0
    flat['max_skill_duration_months'] = max([s.get('duration_months', 0) for s in skills], default=0)

    # Career specific aggregations
    careers = c.get('career_history', [])
    total_job_duration = sum(job.get('duration_months', 0) for job in careers)
    flat['total_job_duration_months'] = total_job_duration

    return flat

def load_data(filepath, sample_size=None):
    records = []
    with open(filepath, 'r') as f:
        for i, line in enumerate(f):
            if sample_size and i >= sample_size:
                break
            records.append(flatten_candidate(json.loads(line)))
    return pd.DataFrame(records)

# Load the first 10,000 rows for quick EDA, or set sample_size=None for all
df = load_data('candidates.jsonl', sample_size=None)
print(f"Loaded {len(df)} records with {len(df.columns)} features.")
df.head()
```

## 2. Missing Value Analysis
Understanding what data is frequently missing is crucial. Missing signals might indicate incomplete profiles, which could be a factor in detecting low-quality or fake accounts.

```python
missing_perc = df.isnull().mean() * 100
missing_perc = missing_perc[missing_perc > 0].sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=missing_perc.values, y=missing_perc.index)
plt.title('Percentage of Missing Values per Feature')
plt.xlabel('% Missing')
plt.show()
```

## 3. Fake Detection Heuristics Analysis
We will investigate several hypotheses that might indicate a profile is fake, keyword-stuffed, or heavily exaggerated.

### Heuristic 1: Logical Contradictions (Experience vs. Job Durations)
Does the `years_of_experience` declared in the profile match the sum of durations in their career history? Exaggerated profiles often claim high years of experience but have short job histories.

```python
df['declared_exp_months'] = df['profile_years_of_experience'] * 12
df['exp_discrepancy'] = df['declared_exp_months'] - df['total_job_duration_months']

plt.figure(figsize=(10, 6))
sns.histplot(df['exp_discrepancy'], bins=50, kde=True)
plt.title('Discrepancy: Declared Experience (months) - Total Job Duration (months)')
plt.xlabel('Discrepancy (Months)')
plt.ylabel('Count')
plt.show()

# Let's identify candidates with massive discrepancies (> 5 years off)
high_discrepancy = df[df['exp_discrepancy'].abs() > 60]
print(f"Candidates with >5 years discrepancy between declared and actual experience: {len(high_discrepancy)} ({len(high_discrepancy)/len(df)*100:.2f}%)")
```

### Heuristic 2: Keyword Stuffing (High Skills but Low Endorsements / Short Duration)
Fake candidates often list 50+ skills but have 0 endorsements or claim to have used them for 0 months. Let's look at the relationship between number of skills and average endorsements.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='num_skills', y='avg_skill_endorsements', data=df, alpha=0.5)
plt.title('Number of Skills vs. Average Endorsements per Skill')
plt.xlabel('Total Number of Skills Listed')
plt.ylabel('Average Endorsements')
plt.show()

keyword_stuffers = df[(df['num_skills'] > 20) & (df['avg_skill_endorsements'] < 1)]
print(f"Potential Keyword Stuffers (>20 skills, <1 avg endorsement): {len(keyword_stuffers)} ({len(keyword_stuffers)/len(df)*100:.2f}%)")
```

### Heuristic 3: Redrob Engagement Signals (Bot-like behavior)
Bots or spam accounts might spam applications but have terrible recruiter response rates and profile completion scores.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(x='signal_applications_submitted_30d', y='signal_recruiter_response_rate', data=df, alpha=0.5, ax=axes[0])
axes[0].set_title('Applications vs Recruiter Response Rate')
axes[0].set_xlabel('Applications Submitted (30 days)')
axes[0].set_ylabel('Recruiter Response Rate')

sns.histplot(df['signal_profile_completeness_score'], bins=20, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Profile Completeness Score')
axes[1].set_xlabel('Completeness Score')

plt.tight_layout()
plt.show()

spam_apps = df[(df['signal_applications_submitted_30d'] > 50) & (df['signal_recruiter_response_rate'] < 0.05)]
print(f"Spammers (>50 applications, <5% response rate): {len(spam_apps)} ({len(spam_apps)/len(df)*100:.2f}%)")
```

### Heuristic 4: Unverified Accounts
Does the account lack basic verification (Email, Phone, LinkedIn) while claiming top tier stats?

```python
df['unverified_count'] = (~df['signal_verified_email']).astype(int) + \
                         (~df['signal_verified_phone']).astype(int) + \
                         (~df['signal_linkedin_connected']).astype(int)

sns.countplot(x='unverified_count', data=df)
plt.title('Number of Unverified Channels (Email, Phone, LinkedIn)')
plt.xlabel('Count of Unverified Channels (0 = Fully Verified, 3 = Completely Unverified)')
plt.show()

completely_unverified = df[df['unverified_count'] == 3]
print(f"Completely Unverified Accounts: {len(completely_unverified)} ({len(completely_unverified)/len(df)*100:.2f}%)")
```

## Conclusion & Next Steps for Stage 1 Filter
Based on the analysis above, we can define a multi-factor **"Trust Score"** or a boolean **"Is_Fake"** filter.

Proposed Rule-out conditions for Stage 1:
1. **`exp_discrepancy`** > Threshold (e.g. 5 years).
2. **`unverified_count`** == 3 (No email, phone, or LinkedIn).
3. **`num_skills`** > Threshold AND **`avg_skill_endorsements`** < Threshold (Keyword stuffing).
4. **`signal_profile_completeness_score`** < Minimum threshold.

Candidates passing this lightweight CPU-bound check will proceed to Stage 2.
